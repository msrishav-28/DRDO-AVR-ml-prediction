```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="darkgrid", context="paper", font_scale=1.2)

# Set Local Paths
BASE_DIR = os.getcwd()
RESULTS_FILE = os.path.join(BASE_DIR, "outputs", "results", "publication_results.json")
OPTUNA_DIR = os.path.join(BASE_DIR, "outputs", "results", "sweeps")

print("Libraries successfully initialized.")
```

### 1. Extract & Visualize Publication Metrics
Parsing the exact Deep Learning AUROC Differentials to prove the significance of the Physics Constraints!

```python
with open(RESULTS_FILE, 'r') as f:
    pub_data = json.load(f)

eval_data = pub_data['evaluation']['held_out']['fault_10s']
models = ['LSTM (Pure Data)', '1D-CNN (Temporal Ablation)', 'PINN (Physics-Informed)']
auroc_means = [eval_data['lstm_auroc_mean'], eval_data['cnn_auroc_mean'], eval_data['pinn_auroc_mean']]
auroc_stds = [eval_data['lstm_auroc_std'], eval_data['cnn_auroc_std'], eval_data['pinn_auroc_std']]

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=models, y=auroc_means, yerr=auroc_stds, capsize=.1, palette='viridis')
plt.title('Performance Supremacy (10-Second Physical Fault Horizon)', pad=20, weight='bold')
plt.ylabel('AUROC Score (Higher is Better)')
plt.ylim(0.2, 0.85)

for i, mean in enumerate(auroc_means):
    ax.text(i, mean + 0.02, f"{mean:.4f}", ha='center', weight='bold')

plt.show()
```

### 2. Verify Optuna GPU Bayesian Iterations
Opening the dynamically generated sequence variables to see which model weights performed best.

```python
import glob

sweep_files = glob.glob(os.path.join(OPTUNA_DIR, "*.csv"))
if sweep_files:
    latest_sweep = max(sweep_files, key=os.path.getctime)
    df_optuna = pd.read_csv(latest_sweep)
    
    print(f"Loaded Optuna Sweep: {os.path.basename(latest_sweep)}\n")
    
    # Display the Best Trial Discovered natively in Markdown
    best_trial = df_optuna.loc[df_optuna['value'].idxmin()]
    print("🔥 BEST SOTA PARAMETER COMBINATION 🔥")
    print(f"- Validation Loss: {best_trial['value']:.4f}")
    print(f"- Learning Rate:   {best_trial['params_lr']:.5f}")
    print(f"- Physics Weight:  {best_trial['params_physics_weight']:.3f}")
    print(f"- Dropout Rate:    {best_trial['params_dropout_rate']:.3f}")
    print(f"- Focal Gamma:     {best_trial['params_focal_loss_gamma']:.3f}")
    
    display(df_optuna[['number', 'value', 'params_physics_weight', 'params_lr', 'duration']])
else:
    print("No sweep data found. Have you executed run_heavy_sweep.py on the GPU yet?")
```
